Linear Regression: Single varialble

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv('lr_prices1.csv')
df
```

```python
%matplotlib inline
plt.scatter(df.area, df.price, color='blue', marker='+')
plt.xlabel('area')
plt.ylabel('price')
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['area']], df.price)
reg.predict([[3700]])
```

```python
y = (reg.coef_)*3700 + reg.intercept_ #y=mx+c
y
```

```python
d = [1000, 1500, 2300, 3300, 3500, 4120, 4480, 5120, 5780, 4810, 8500, 7120, 6630, 9700, 8190]
d = pd.DataFrame(d,columns=['area'])
d.to_csv('area1.csv')
```

```python
d = pd.read_csv('area1.csv')
d = d[['area']]
d
```

```python
p = reg.predict(d)
```

```python
d['prices'] = p
```

```python
d.to_csv('price_predicted1.csv')
```

```python
%matplotlib inline

plt.scatter(df.area, df.price, color='green', marker='+')
plt.xlabel('area', fontsize=20)
plt.ylabel('price', fontsize=20)
plt.plot(df.area, reg.predict(df[['area']]), color='blue')
```

```python
%matplotlib inline

plt.scatter(d.area, d.prices, color='green', marker='+')
plt.xlabel('area', fontsize=20)
plt.ylabel('prices', fontsize=20)
plt.plot(d.area, reg.predict(d[['area']]), color='blue')
```
